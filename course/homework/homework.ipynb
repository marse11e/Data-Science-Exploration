# Определение вида треугольника

**Цель:**  
Написать программу на Python, которая будет проверять, могут ли три заданных числа быть сторонами треугольника, и если могут, определит вид этого треугольника по его сторонам и углам.

### Условие задачи:
1. Напишите программу, которая принимает на вход три числа \(a\), \(b\) и \(c\), представляющие длины сторон треугольника.
2. Программа должна сначала проверить, могут ли эти числа образовать треугольник. Для этого используйте условие:
   - Сумма любых двух сторон должна быть больше третьей.
   
   Если хотя бы одно из этих условий не выполняется, программа должна вывести: **"Это не треугольник"**.

3. Если треугольник существует, программа должна определить его тип:
   - **Равносторонний треугольник**, если все стороны равны.
   - **Равнобедренный треугольник**, если две стороны равны.
   - **Разносторонний треугольник**, если все стороны разные.

4. Программа также должна определить вид треугольника по углам:
   - **Прямоугольный треугольник**, если выполняется теорема Пифагора для одной из сторон.
   - **Остроугольный треугольник**, если все углы острые.
   - **Тупоугольный треугольник**, если один угол тупой.

### Подсказки:
1. Для проверки треугольника по углам используйте теорему Пифагора:
   - Прямоугольный треугольник: \(a^2 + b^2 = c^2\)
2. Для проверки тупоугольного треугольника убедитесь, что сумма квадратов двух меньших сторон меньше квадрата третьей стороны.

### Задание:
1. Напишите программу, используя предложенные шаги.
2. Протестируйте её с разными значениями сторон, чтобы проверить, как программа определяет разные виды треугольников.

````python
a, b, c = 10, 10, 10

if a + b > c and a + c > b and b + c > a:
    print("Это треугольник")
    print("Равносторонний" if a == b == c else "Равнобедренный" if a == b or a == c or b == c else "Разносторонний")
    print("Прямоугольный" if a**2 + b**2 == c**2 or a**2 + c**2 == b**2 or b**2 + c**2 == a**2 else 
          "Остроугольный" if a**2 + b**2 > c**2 and a**2 + c**2 > b**2 and b**2 + c**2 > a**2 else "Тупоугольный")
else:
    print("Это не треугольник")
````

### Задача: Поиск минимального числа в списке

**Описание:** 

Напишите функцию, которая принимает список чисел и возвращает минимальное число из этого списка. Функция должна работать для списков любого размера и содержать как положительные, так и отрицательные числа.

**Требования:**

1. Функция должна называться `find_minimum`.
2. Она должна принимать один аргумент — список чисел (например, `[3, 5, -1, 7, 0]`).
3. Функция должна возвращать минимальное число из переданного списка.
4. Если список пустой, функция должна возвращать `None` или выводить сообщение об ошибке.


### Подсказка:

Вы можете использовать цикл для перебора элементов списка или встроенную функцию `min()` для нахождения минимального значения. Также не забудьте обработать случай, когда список пуст. 

### Ожидаемый результат:

При вызове функции с непустым списком чисел функция должна вернуть минимальное число, а при вызове с пустым списком — `None` или соответствующее сообщение.

````python
def find_minimum(numbers):
    return min(numbers) if numbers else None

print(find_minimum([3, 5, -1, 7, 0])) 
````

### Задача: Повторение строки

**Описание:**

Напишите функцию, которая принимает строку и целое число, указывающее, сколько раз строка должна быть повторена. Функция должна возвращать новую строку, состоящую из повторяющейся строки, разделённой пробелами.

**Требования:**

1. Функция должна называться `repeat_string`.
2. Она должна принимать два аргумента:
   - `text` (строка), которую нужно повторить.
   - `count` (целое число), указывающее количество повторений.
3. Функция должна возвращать строку, содержащую `count` повторений `text`, разделённых пробелами.
4. Если `count` равен нулю или меньше, функция должна возвращать пустую строку.


### Ожидаемый результат:

При вызове функции с указанной строкой и количеством повторений функция должна вернуть строку, состоящую из повторяющихся значений, разделённых пробелами. Если количество повторений равно нулю, функция должна вернуть пустую строку.

````python
def repeat_string(text, count):
    return " ".join([text] * count) if count > 0 else ""

print(repeat_string("hello", 3)) 
````

## Задача: Написать функцию для извлечения домена из списка email-адресов.

### Условия задачи:
1. Вам дан список email-адресов, например: `['dhdhcv@gmail.com', 'admin@test.org', 'user@example.com']`.
2. Каждый email состоит из двух частей:
   - Логин (часть до символа `@`);
   - Домен (часть после символа `@`).
3. Необходимо написать функцию с использованием lambda, которая для каждого email-адреса в списке будет возвращать только домен (часть после символа `@`).

### Ожидаемый результат:
```python
['gmail.com', 'test.org', 'example.com']
```

````python
emails = [
    'john.doe@gmail.com',
    'admin@company.org',
    'user123@yahoo.com',
    'support@website.net',
    'info@business.co',
    'client@marketplace.com',
    'customer@store.net',
    'boss@enterprise.biz',
    'dev@developer.io',
    'employee@corporation.com',
    'leader@organization.org'
]

def get_domain(email):
    at_index = email.find('@')
    return email[at_index+1:]

print(list(map(get_domain, emails)))
print(list(map(lambda email: email[email.find('@')+1:], emails)))
````

5. **Список ключей и значений:** В последней части задачи нужно создать словарь, а затем вывести все ключи и значения, используя встроенные методы Python.

````python
def create_dict(keys, values):
    return {keys[i]: values[i] for i in range(len(keys))}

def print_keys_and_values(dictionary):
    print("Keys:", list(dictionary.keys()))
    print("Values:", list(dictionary.values()))

print_keys_and_values(create_dict(['name', 'age', 'city'], ['Marselle', 21, 'Almaty']))
````

Задача "Проверка на простое число" заключается в том, чтобы написать функцию, которая проверяет, является ли заданное целое число простым.

**Условия задачи:**
- Простое число — это число больше 1, которое делится только на себя и на 1.
- Задача состоит в том, чтобы определить, является ли число простым.

**Входные данные:**
- Одно целое число \( n \), где \( n \geq 0 \).

**Выходные данные:**
- Булево значение: `True`, если число простое, и `False`, если не простое.

**Пример:**
1. Ввод: `5` → Вывод: `True` (5 простое число, так как делится только на 1 и 5).
2. Ввод: `4` → Вывод: `False` (4 не простое, так как делится на 2).

**Цель задачи:** Оптимально определить, простое ли число, избегая лишних проверок.

````python
def is_prime(n):
    if n <= 1:
        return 'Не являются простым'

    if n == 2:
        return 'Простое число'

    if n % 2 == 0:
        return 'Не являются простым'

    for i in range(3, int(n ** 0.5) + 1, 2):
        if n % i == 0:
            return 'Не являются простым'
    return 'Простое число'


def is_prime_2(n):
    if n <= 1 or (n > 2 and n % 2 == 0):
        return 'Не являются простым'
    
    if all(n % i for i in range(3, int(n ** 0.5) + 1, 2)) or n == 2:
        return 'Простое число'
    
    return 'Не являются простым'


# print(is_prime_2(2))


# Сокращенный код для нахождения простого числа
n = int(input('Введите число: '))
print('Простое число' if all(n % i for i in range(3, int(n ** 0.5) + 1, 2)) or n == 2 else 'Не являются простым')
````
